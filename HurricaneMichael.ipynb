```python
import pandas as pd
import json

import folium as fm
from folium.plugins import MarkerCluster
```

```python
#dataset of addresses in Finland
location = "datasets/datasets_hurricaneMichael.csv"

#load data into dataframe
#seperator between values in file is a semicolon
hurr_df = pd.read_csv(location)
hurr_df.head()
```

```python
hurr_df['Long'] = 0 - hurr_df['Long']
```

```python
hurr_df.head()
```

```python
#only include rows with non-null latitudes (which means longitude is also not null) and non-null mass
mhurr_df = hurr_df.loc[(hurr_df['Lat'].notnull())]
mhurr_df.count()
```

```python
mhurr_df
```

```python
m = fm.Map([43, -100], tiles='Stamen Terrain', zoom_start=4)

for index, row in mhurr_df.iterrows(): 
    fm.Marker(location=[row['Lat'],row['Long']], popup = None , tooltip = "{0} MHP Winds".format(row['Wind'])).add_to(m)

m
```
