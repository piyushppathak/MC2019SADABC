```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
#from nltk.tokenize import sent_tokenize
#from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
#from nltk.stem import WordNetLemmatizer
#from nltk.sentiment.vader import SentimentIntensityAnalyzer

from string import punctuation
```

```python
#[create an empty list here to hold the tokens at the end]
wrds = []

with open('datasets/12dancingprincesses.txt', 'r') as f:
    for line in f:
        cline = line.strip() #get rid of newline character

        if cline == '': pass  #this will skip over lines that only had a newline and are now blank
        else:
            tknls = word_tokenize(cline.lower()) #[write the function to tokenize the words]

            for token in tknls:
               #[write the function to append each token to the empty list you created at the start of this code]
               if token not in punctuation:
                   wrds.append(token)
```

```python
#list of english stopwords
eng_stopwords = stopwords.words('english')
eng_stopwords
```

```python
rm_count = 0
new_words = []  #list to hold new words

for word in wrds:
    if word not in eng_stopwords:
        new_words.append(word)
    else: rm_count += 1
```

```python
fd_nw = FreqDist(new_words)
fd_nw.most_common(10)
```

```python
unwanted_valu = {'’', '‘'} 
```

```python
cleaned_words = [nwelmnt for nwelmnt in new_words if nwelmnt not in unwanted_valu]
```

```python
fd_nw = FreqDist(cleaned_words)
fd_nw.most_common(10)
```
